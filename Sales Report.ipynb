### CLEANING MY DATASET USING PYTHON


```python
import pandas as pd
import numpy as np
```

```python
# Reading in my dataset
df = pd.read_csv("supermarket_sales - Sheet1.csv")
df
```

```python
# To know my exact number of rows and columns
df.shape
```

```python
# checking the datatypes
df.dtypes
```

```python
# CORRECTING THE DATA TYPE FORMAT IN DATE COLUMN TO A DATETIME FORMAT
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
df.dtypes
```

```python
# checking for null values
df.isnull().sum()
```

```python
### checking for duplicates
df.duplicated().sum()
```

```python
# Renaming column cogs and payment
df.rename(columns={"cogs":"Cost of goods sold"}, inplace=True)
df.rename(columns={"Payment":"Payment Type"}, inplace=True)
```

```python
df
```

```python
# Removing columns
df.drop(columns=["Invoice ID", "Time"], inplace=True)
```

```python
df
```

```python
# Capitalizing every word in each column
df.columns = df.columns.str.upper()
```

```python
df
```

```python
# converting date column into day, month and year and creating seperate columns for it
df["DAY"] = df["DATE"].dt.day_name()
df["MONTH"] = df["DATE"].dt.month_name()
df["YEARS"] = df["DATE"].dt.year
```

```python
df
```

```python
# Saving my dataset after cleaning
df.to_csv('Cleaned_Sales_Report_Data.csv', index=False)
```
